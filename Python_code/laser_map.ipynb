```python
import pandas as pd
import numpy as np
import csv
import math
```

```python
df = pd.read_csv('./data/beam_23.csv')
df
```

```python
#maximum
max_v = df.max().max()
max_v

#Change column names
len_num = len(df.columns)
v_list = list(range(0, len_num))
df.columns=v_list
df
```

```python
#Put the first line in the background
df_t = df.T
mean_val = df_t[0].mean()
mean_val
```

```python
#Find the column containing the largest value
col_des = df.describe()
col_max = col_des.T
a = col_max["max"].idxmax()
```

```python
df_max = df[a]
df_max = pd.DataFrame(df_max)
df_max
```

```python
df_max.columns=["strength"]
df_max["y_pos"] = list(range(0, len(df_max)))
#Pixel size 5.5Î¼m
df_max["width"] = df_max["y_pos"] * 0.0055 *2
df_max["relative_intensity"] = df_max["strength"]/max_v
```

```python
df_max
```

```python
#Calculate laser diameter
df_cal = df_max[df_max["relative_intensity"]>= (1/math.exp(2))]
laser_rad = len(df_cal)*0.0055 *2
laser_rad
```

```python
df_max.to_csv("./data/df_max.csv", sep=",")
```
