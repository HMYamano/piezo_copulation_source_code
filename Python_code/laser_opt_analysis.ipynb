```python
def fast_concat(dfs):
    def fast_flatten(input_list):
        return list(chain.from_iterable(input_list))

    col_names = dfs[0].columns
    df_dict = dict.fromkeys(col_names, [])
    for col in col_names:
        extracted = (d[col] for d in dfs)
        df_dict[col] = fast_flatten(extracted)

    df = pd.DataFrame.from_dict(df_dict)[col_names]
    return df
```

```python
def cal_length(df):
    colname = ["frame", "M_rchex", "M_rchey", "M_lchex", "M_lchey", "length", "Class"]
    df_reslts = pd.DataFrame()
    df_total = pd.DataFrame()
    i = 1
    cl = 0
    df['Class'].astype(str)
    maxclass = df.at[df.index[(len(df)-1)], "Class"]
    for i in range(int(maxclass)):
        n = 0
        cl = cl + 1
        total_length = 0
        df_cal = df[df["Class"] == str(cl)]
        df_cal = df_cal.reset_index(drop=True)
        rx_pre = df_cal.loc[0, "M_rchex"] 
        ry_pre = df_cal.loc[0, "M_rchey"] 
        lx_pre = df_cal.loc[0, "M_lchex"] 
        ly_pre = df_cal.loc[0, "M_lchey"]
        for j in range(len(df_cal)):
            
            frame = j
            rx_pos = df_cal.at[df_cal.index[n], "M_rchex"] 
            ry_pos = df_cal.at[df_cal.index[n], "M_rchey"]
            lx_pos = df_cal.at[df_cal.index[n], "M_lchex"] 
            ly_pos = df_cal.at[df_cal.index[n], "M_lchey"]
            
            length = math.sqrt(((rx_pos + lx_pos )/2 - (rx_pre + lx_pre)/2)**2 + ((ry_pos + ly_pos)/2 - (ry_pre + ly_pre)/2)**2)

            dfs = pd.DataFrame([[frame, rx_pos, ry_pos, lx_pos, ly_pos, length, cl]], columns = colname)
            df_reslts = pd.concat([df_reslts, dfs])
            n = n + 1
            rx_pre = rx_pos
            ry_pre = ry_pos
            lx_pre = lx_pos
            ly_pre = ly_pos
            total_length = total_length + length
        dfss = pd.DataFrame([[cl, total_length]], columns = ["Class", "Total_length"])
        df_total = pd.concat([df_total, dfss])

    df_reslts = df_reslts.reset_index(drop=True)
    df_total = df_total.reset_index(drop=True)
    return df_reslts, df_total
```

```python
df_atr, df_atr_total = cal_length(df_atr_raw)
df_etoh, df_etoh_total = cal_length(df_etoh_raw)
```
